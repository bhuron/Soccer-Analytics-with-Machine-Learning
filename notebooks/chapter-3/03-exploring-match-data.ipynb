## Setup and Imports

```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the match data (from previous notebook)
BASE = Path("open-data/data")
matches = json.load(open(BASE / "matches/72/30.json", "r"))
matches_df = pd.json_normalize(matches)

print(f"Loaded {len(matches_df)} matches")
```

## Numerical Features: The Quantifiable Game

Numerical features are the measurable quantities in our dataset. They can be:
- **Continuous**: Can take any value within a range (e.g., pass length, shot distance)
- **Discrete**: Only specific integer values (e.g., number of goals, shots taken)

### Continuous Variables: Goals Scored Distribution

```python
# Create a histogram of home and away goals
plt.figure(figsize=(10, 6))
sns.histplot(matches_df["home_score"], bins=range(0, 8), alpha=0.7, label="Home", color="blue")
sns.histplot(matches_df["away_score"], bins=range(0, 8), alpha=0.7, label="Away", color="red")
plt.xlabel("Goals Scored")
plt.ylabel("Frequency")
plt.title("Distribution of Goals Scored (Home vs Away)")
plt.legend()
plt.show()
```

**What this tells us:**
- Most matches see one or two goals scored by each team
- Confirms soccer's reputation as a low-scoring sport
- Home and away distributions are similar (no strong home advantage in this dataset)
- Understanding typical goal ranges helps set expectations for performance evaluation

### Discrete Variables: Match Duration

```python
# Distribution of match duration
plt.figure(figsize=(8, 5))
sns.histplot(matches_df["match_duration"], bins=20)
plt.xlabel("Match Duration (minutes)")
plt.ylabel("Count")
plt.title("Distribution of Match Duration")
plt.show()
```

**What this tells us:**
- Most matches cluster around 90-95 minutes
- Some extend longer due to stoppage time or extra time in knockout rounds
- Anomalies stand out immediately in this visualization

## Categorical Features: The Qualitative Aspects

Categorical features represent qualitative data without mathematical ordering. They answer "what" and "who" questions rather than "how much."

Examples include:
- Team names
- Player names
- Event types (Pass, Shot, Foul)
- Match outcomes (Win, Draw, Loss)

### Exploring Team Participation

```python
# Count appearances for each team
team_counts = pd.concat([
    matches_df["home_team.home_team_name"],
    matches_df["away_team.away_team_name"]
]).value_counts()

# Create a horizontal bar plot
plt.figure(figsize=(10, 8))
sns.barplot(y=team_counts.index[:10], x=team_counts.values[:10], orient='h')
plt.xlabel("Number of Matches")
plt.title("Top 10 Teams by Number of Matches Played")
plt.tight_layout()
plt.show()
```

**What this tells us:**
- Teams with more matches reached later tournament stages
- Teams with fewer matches were eliminated earlier
- This simple count reveals tournament progression without detailed match results
- Quick way to identify the most successful teams

## Ordinal Features: The Ordered Categories

Ordinal features are categorical but have a clear, meaningful order. They're like categorical data with a ranking system built in.

Examples in soccer:
- Competition stages (Group Stage < Round of 16 < Quarter-finals < Semi-finals < Final)
- Player ratings (Poor < Average < Good < Excellent)
- Match importance levels

### Tournament Progression Analysis

```python
# Define the correct order of tournament stages
stage_order = [
    "Group Stage",
    "Round of 16",
    "Quarter-finals",
    "Semi-finals",
    "Third Place Play-off",
    "Final"
]

# Convert to ordered categorical
matches_df["competition_stage.name"] = pd.Categorical(
    matches_df["competition_stage.name"],
    categories=stage_order,
    ordered=True
)

# Count matches per stage
stage_counts = matches_df["competition_stage.name"].value_counts().sort_index()

# Create the visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=stage_counts.values, y=stage_counts.index, orient='h')
plt.xlabel("Number of Matches")
plt.title("Number of Matches per Tournament Stage")
plt.tight_layout()
plt.show()
```

**What this tells us:**
- The ordered visualization follows the natural progression of the tournament
- Group Stage has the most matches (36)
- Followed by Round of 16 (8 matches), and so on
- Respects the tournament's knockout structure
- Makes the data much easier to interpret than alphabetical ordering

## Summary

In this notebook, we explored three types of features in soccer data:

1. **Numerical Features**: Quantifiable measurements like goals and match duration
   - Continuous: Can take any value in a range
   - Discrete: Only specific integer values

2. **Categorical Features**: Qualitative data like team names and outcomes
   - No inherent ordering
   - Answer "what" and "who" questions

3. **Ordinal Features**: Categorical data with meaningful order
   - Tournament stages, ratings, importance levels
   - Ordering matters for interpretation

Each type requires different visualization and analysis techniques. Understanding these differences is crucial for effective exploratory data analysis.
