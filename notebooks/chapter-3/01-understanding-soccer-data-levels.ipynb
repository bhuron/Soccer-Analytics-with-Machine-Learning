## Introduction

Before we dive into the data, let's talk about how it's organized. Think of it like watching a game from different perspectives. You could be in a helicopter seeing the whole stadium, on the sideline watching the teams, or right on the pitch following a single player. Soccer data has a similar layered structure, and understanding these levels is crucial because the kinds of questions you can answer depend on the level of data you have.

## Match-Level Data

This is the 30,000-foot view. It gives you the final score, the teams that played, the competition name, and the date. It's the kind of information you'd find on a scoreboard or fixture list. Match-level data can describe a single match or be aggregated across many matches—for example, all games in the group stage of a tournament.

### What Questions Can Match-Level Data Answer?

Match-level data is perfect for answering high-level questions like:
- Who won the most games in the group stage?
- Which teams scored the most goals in the tournament?
- How many matches went to extra time?

### Limitations

You can't analyze a team's passing patterns with only match-level data, but you can identify trends across competitions and seasons. It's the foundation for tournament analysis and league standings.

## Player-Level Data

Now we're on the sideline. This level tracks the performance of individual players across a match or tournament. How many goals did Megan Rapinoe score? How many kilometers did Sam Kerr run? How accurate were Alex Morgan's passes?

Player-level data is often aggregated from finer-grained event data and helps us compare players across different roles and teams.

### What Questions Can Player-Level Data Answer?

It's essential for:
- Player recruitment and scouting
- Performance evaluation and ratings
- Fantasy soccer and player awards
- Understanding individual contributions to team success

## Event-Level Data

This is the most granular view, putting you right on the pitch with the players. It records every individual event in a match—each pass, shot, tackle, dribble, or foul—along with who did it, where it happened on the pitch, and when.

**Important Note**: In soccer analytics, the term "event" refers specifically to any discrete on-ball action that changes or continues possession. It doesn't mean "tournament" or "competition event," as in everyday English.

### What Questions Can Event-Level Data Answer?

Event-level data allows analysts to study:
- Tactical patterns and team formations
- Player actions and decision-making
- The flow and momentum of matches
- Spatial control and territorial dominance

### The Power of Event-Level Data

The beauty of event-level data is that you can **aggregate it upward** to create player-level or match-level summaries—but not the other way around. Once you have event data, you can answer questions at any level of detail.

## Summary: Choosing the Right Data Level

Understanding these levels is crucial because each connects to a different class of analytical questions. You can't determine a tournament winner from a single pass, just as you can't analyze a team's passing structure with only match-level data.

In this book, we'll work with all three levels, often combining them to build a complete analytical picture.

### Quick Reference Table

| Data Level | Granularity | Example Questions | Can Aggregate To |
|------------|-------------|-------------------|------------------|
| Match | Highest | Who won? What was the score? | Tournament/Season stats |
| Player | Medium | How many goals did Player X score? | Team stats, Match stats |
| Event | Lowest | Where did each pass go? | Player stats, Match stats |
