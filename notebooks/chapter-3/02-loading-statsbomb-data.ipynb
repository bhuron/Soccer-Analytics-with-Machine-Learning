## Setup and Imports

## Getting the Data

StatsBomb is a well-regarded provider in the soccer analytics community, and their open release is one of the most detailed event datasets available: every pass, shot, dribble, foul, and player position is carefully logged.

The StatsBomb data lives on GitHub. To download it, open your command line and type:

```bash
git clone https://github.com/statsbomb/open-data.git
```

This command fetches the full history of StatsBomb's open data. Depending on your connection, it can take ten minutes or more.

### Data Structure

Once downloaded, you'll see a folder called `open-data` with several subfolders. The most important one is `data/`, which contains the actual JSON files organized by competition and season.

For the Women's World Cup 2019, the matches live in:
```
open-data/data/matches/72/30.json
```

Where:
- `72` is the competition ID for the Women's World Cup
- `30` is the season ID for 2019
- `.json` indicates JavaScript Object Notation format

## Loading Match Data

With the files on disk, the next step is to load them into a pandas DataFrame. A DataFrame is essentially a smart table: it combines the familiarity of a spreadsheet with the power of a database query.

````python
# Define the base path (adjust if needed based on where you cloned the repo)
BASE = Path("open-data/data")

# Load all WWC 2019 matches
matches = json.load(open(BASE / "matches/72/30.json", "r"))
matches_df = pd.json_normalize(matches)

print(f"Number of matches: {len(matches_df)}")
matches_df.head()
````

The `pd.json_normalize` function is particularly useful here—it takes the nested JSON structure and flattens it into a clean DataFrame with columns like `match_name`, `competition_stage_name`, and `season_name`.

## First Look at the Data

Before a coach sets up tactics, they scan the team sheet to see who is available. Our equivalent is a quick inspection of the DataFrame: what columns do we have, what types of values, and where are the gaps?

````python
# Basic structure of the matches DataFrame
matches_df.info()
````

The `info()` method reports:
- Number of rows
- Column names
- Data types (integer, string, float)
- Non-null counts (to identify missing data)

````python
# Quick descriptive stats (mainly for numeric columns)
matches_df.describe()
````

The `describe()` method summarizes numeric columns with counts, averages, and ranges.

## Exploring Key Columns

Let's examine some of the most important columns in our matches DataFrame.

````python
# Display a subset of relevant columns
key_columns = [
    'match_id',
    'match_date',
    'home_team.home_team_name',
    'away_team.away_team_name',
    'home_score',
    'away_score',
    'competition_stage.name'
]

# Check which columns actually exist (column names may vary)
available_columns = [col for col in key_columns if col in matches_df.columns]
matches_df[available_columns].head(10)
````

## Understanding Tournament Structure

````python
# Check tournament stages
if 'competition_stage.name' in matches_df.columns:
    print("Tournament stages:")
    print(matches_df['competition_stage.name'].value_counts())
else:
    print("Stage information not found in expected column")
    print("\nAvailable columns:")
    print([col for col in matches_df.columns if 'stage' in col.lower()])
````
