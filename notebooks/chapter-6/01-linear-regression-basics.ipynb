## The Intuition: Finding the Line of Best Fit\n\n**Linear Regression** finds a straight-line relationship between input variables (features) and an output variable (target).\n\n**Think of it like a scout estimating player value:**\n- More goals scored → Higher market value\n- More assists → Higher market value\n- Younger age → Potentially higher value\n\nLinear regression does this mathematically by finding the **best straight line** through the data.\n\n**The equation:** `Market Value = slope × Goals + intercept`

## Load Data: Player Market Values\n\nWe'll use simulated data representing attacking players in a top European league.

## Build the Linear Regression Model\n\nNow let's build our model using scikit-learn.

```python
# Prepare features (X) and target (y)\nX = df[['Goals']]  # Must be 2D array\ny = df['MarketValue_Millions']\n\n# Create and train the model\nmodel = LinearRegression()\nmodel.fit(X, y)\n\n# Get model parameters\nintercept = model.intercept_\nslope = model.coef_[0]\n\nprint(\"Model trained successfully!\")\nprint(f\"\\nModel Parameters:\")\nprint(f\"  Intercept: {intercept:.2f}\")\nprint(f\"  Slope (coefficient for Goals): {slope:.2f}\")\nprint(f\"\\nModel equation: Market Value = {slope:.2f} × Goals + {intercept:.2f}\")
```

## Interpret the Coefficients\n\n**Slope (≈4.15):** For every additional goal scored, a player's market value increases by approximately €4.15 million.\n\n**Intercept (≈-5.5):** The baseline value when goals = 0. In this case, it's negative, which doesn't make real-world sense. This is common in simple models and highlights that the intercept is often just a mathematical necessity to position the line correctly.\n\n**Soccer Insight:** The slope quantifies the relationship between goal-scoring ability and market value. This could help clubs make data-driven transfer decisions!

## Evaluate Model Performance: R-squared\n\n**R-squared (R²)** tells us what percentage of variation in market value can be explained by goals scored.\n\n- R² = 1: Perfect fit (all points on the line)\n- R² = 0: Model explains nothing\n- R² = 0.95: Model explains 95% of variation

```python
# Get predictions\ny_pred = model.predict(X)\n\n# Calculate R-squared\nr2 = r2_score(y, y_pred)\n\nprint(f\"R-squared: {r2:.3f}\")\nprint(f\"\\nInterpretation: {r2*100:.1f}% of the variation in market values\")\nprint(f\"can be explained by the number of goals scored.\")\n\nif r2 > 0.9:\n    print(\"\\nThis is an excellent fit! (Though unusually high for real-world data)\")
```

## Residual Analysis\n\n**Residuals** are the differences between actual and predicted values. A good model should have residuals randomly scattered around zero.

```python
# Calculate residuals\nresiduals = y - y_pred\n\nplt.figure(figsize=(10, 6))\nsns.scatterplot(x=y_pred, y=residuals, s=100)\nplt.axhline(y=0, color='r', linestyle='--', linewidth=2)\nplt.xlabel('Predicted Market Value')\nplt.ylabel('Residuals')\nplt.title('Residual Plot')\nplt.tight_layout()\nplt.show()\n\nprint(\"If residuals show a pattern (e.g., U-shape), linear regression may not be the best fit.\")\nprint(\"Random scatter around zero = good model!\")
```
