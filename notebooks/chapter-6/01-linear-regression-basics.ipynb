```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
```

## The Intuition: Finding the Line of Best Fit

**Linear Regression** finds a straight-line relationship between input variables (features) and an output variable (target).

**Think of it like a scout estimating player value:**
- More goals scored → Higher market value
- More assists → Higher market value
- Younger age → Potentially higher value

Linear regression does this mathematically by finding the **best straight line** through the data.

**The equation:** `Market Value = slope × Goals + intercept`


## Load Data: Player Market Values

We'll use simulated data representing attacking players in a top European league.


```python
# Simulated data for player market values
data = {
    'Player': ['Player A', 'Player B', 'Player C', 'Player D', 'Player E', 'Player F', 'Player G'],
    'Goals': [22, 15, 8, 25, 12, 18, 5],
    'MarketValue_Millions': [90, 60, 25, 100, 45, 70, 15]
}
df = pd.DataFrame(data)

print(\"Player Market Value Dataset:\")
print(df)
print(f\"\nDataset shape: {df.shape}\")
print(f\"Average market value: €{df['MarketValue_Millions'].mean():.1f}M\")
```

## Visualize the Relationship

Before building a model, always visualize! A scatter plot shows the relationship between goals and market value.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Goals', y='MarketValue_Millions', data=df, s=100)
plt.title('Player Market Value vs. Goals Scored', fontsize=14)
plt.xlabel('Goals Scored Last Season')
plt.ylabel('Market Value (in Millions €)')
plt.tight_layout()
plt.show()

print(\"Observation: There's a clear positive relationship!\")
print(\"Players who score more goals tend to have higher market values.\")
```

## Build the Linear Regression Model

Now let's build our model using scikit-learn.


```python
# Prepare features (X) and target (y)
X = df[['Goals']]  # Must be 2D array
y = df['MarketValue_Millions']

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Get model parameters
intercept = model.intercept_
slope = model.coef_[0]

print(\"Model trained successfully!\")
print(f\"\nModel Parameters:\")
print(f\"  Intercept: {intercept:.2f}\")
print(f\"  Slope (coefficient for Goals): {slope:.2f}\")
print(f\"\nModel equation: Market Value = {slope:.2f} × Goals + {intercept:.2f}\")
```

## Interpret the Coefficients

**Slope (≈4.15):** For every additional goal scored, a player's market value increases by approximately €4.15 million.

**Intercept (≈-5.5):** The baseline value when goals = 0. In this case, it's negative, which doesn't make real-world sense. This is common in simple models and highlights that the intercept is often just a mathematical necessity to position the line correctly.

**Soccer Insight:** The slope quantifies the relationship between goal-scoring ability and market value. This could help clubs make data-driven transfer decisions!


## Visualize the Model

Let's plot the regression line on our scatter plot.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Goals', y='MarketValue_Millions', data=df, s=100, label='Actual Data')

# Plot the regression line
plt.plot(X, model.predict(X), color='red', linewidth=3, label='Regression Line')

plt.title('Linear Regression: Market Value vs. Goals Scored', fontsize=14)
plt.xlabel('Goals Scored Last Season')
plt.ylabel('Market Value (in Millions €)')
plt.legend()
plt.tight_layout()
plt.show()

print(\"The red line represents our model's predictions.\")
print(\"It minimizes the overall distance to all data points.\")
```

## Evaluate Model Performance: R-squared

**R-squared (R²)** tells us what percentage of variation in market value can be explained by goals scored.

- R² = 1: Perfect fit (all points on the line)
- R² = 0: Model explains nothing
- R² = 0.95: Model explains 95% of variation


```python
# Get predictions
y_pred = model.predict(X)

# Calculate R-squared
r2 = r2_score(y, y_pred)

print(f\"R-squared: {r2:.3f}\")
print(f\"\nInterpretation: {r2*100:.1f}% of the variation in market values\")
print(f\"can be explained by the number of goals scored.\")

if r2 > 0.9:
    print(\"\nThis is an excellent fit! (Though unusually high for real-world data)\")
```

## Residual Analysis

**Residuals** are the differences between actual and predicted values. A good model should have residuals randomly scattered around zero.


```python
# Calculate residuals
residuals = y - y_pred

plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred, y=residuals, s=100)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Predicted Market Value')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.tight_layout()
plt.show()

print(\"If residuals show a pattern (e.g., U-shape), linear regression may not be the best fit.\")
print(\"Random scatter around zero = good model!\")
```

## Make Predictions

Now we can predict market values for new players!


```python
# Predict for new players
new_players = pd.DataFrame({
    'Goals': [10, 20, 30]
})

predictions = model.predict(new_players)

print(\"Market Value Predictions:\")
for goals, value in zip(new_players['Goals'], predictions):
    print(f\"  Player with {goals} goals: €{value:.1f}M\")
```
