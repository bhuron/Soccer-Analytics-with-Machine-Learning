## 1. Conditional Statements

Conditional statements allow your code to make decisions based on data.

### Example: Calculating Match Points

```python
# Determine points based on match result
team_goals = 2
opponent_goals = 1

if team_goals > opponent_goals:
    points = 3  # Win
    result = "Win"
elif team_goals == opponent_goals:
    points = 1  # Draw
    result = "Draw"
else:
    points = 0  # Loss
    result = "Loss"

print(f"Result: {result}, Points: {points}")
```

## 2. Loops

Loops allow you to process collections of data efficiently.

### For Loops: Iterating Through Collections

```python
# Calculate total points from a list
points_list = [3, 1, 0, 3, 3]  # Points from 5 matches
total_points = 0

for points in points_list:
    total_points += points

print(f"Total points accumulated: {total_points}")
```

### Using zip() to Process Multiple Lists

```python
# Calculate points from goals scored and conceded
team_goals = [2, 1, 3, 0, 2]
opponent_goals = [1, 2, 1, 0, 1]
total_points = 0

for goals, opponent in zip(team_goals, opponent_goals):
    if goals > opponent:
        total_points += 3  # Win
    elif goals == opponent:
        total_points += 1  # Draw
    # else: 0 points for loss

print(f"Total standing points: {total_points}")
```

## 3. List Comprehensions

List comprehensions provide a concise way to create lists and transform data.

```python
# Traditional loop approach
squares = []
for x in range(10):
    squares.append(x**2)

print(f"Squares (loop): {squares}")

# List comprehension approach (more Pythonic)
squares = [x**2 for x in range(10)]
print(f"Squares (comprehension): {squares}")
```

### Soccer Example: Identifying Home Wins

```python
# Determine which matches were home wins
home_goals = [2, 1, 3, 0, 2]
away_goals = [1, 2, 1, 0, 1]

home_wins = [home > away for home, away in zip(home_goals, away_goals)]
print(f"Home wins: {home_wins}")
print(f"Number of home wins: {sum(home_wins)}")
```

## 4. Functions

Functions encapsulate reusable logic, making your code more organized and maintainable.

### Example: Calculate Points Function

```python
def calculate_points(games: list) -> int:
    """
    Calculate the total points from a list of game results.
    
    Arguments:
        games (list of tuples): Each tuple contains (goals_scored, goals_lost)
    
    Returns:
        int: Total points (3 for win, 1 for draw, 0 for loss)
    """
    points = sum(3 for scored, lost in games if scored > lost)
    points += sum(1 for scored, lost in games if scored == lost)
    return points

# Example usage
games = [(2, 1), (1, 1), (3, 0), (0, 2), (2, 1)]
total_points = calculate_points(games)
print(f"Total points accumulated: {total_points}")
```

### Example: Goal Difference Function

```python
def calculate_goal_difference(games: list) -> int:
    """
    Calculate total goal difference from a list of games.
    
    Arguments:
        games (list of tuples): Each tuple contains (goals_scored, goals_lost)
    
    Returns:
        int: Total goal difference
    """
    return sum(scored - lost for scored, lost in games)

# Example usage
games = [(2, 1), (1, 1), (3, 0), (0, 2), (2, 1)]
goal_diff = calculate_goal_difference(games)
print(f"Goal difference: {goal_diff:+d}")  # :+d shows + sign for positive numbers
```

### Example: Team Statistics Function

```python
def calculate_team_stats(games: list) -> dict:
    """
    Calculate comprehensive team statistics.
    
    Arguments:
        games (list of tuples): Each tuple contains (goals_scored, goals_lost)
    
    Returns:
        dict: Dictionary with wins, draws, losses, points, and goal difference
    """
    wins = sum(1 for scored, lost in games if scored > lost)
    draws = sum(1 for scored, lost in games if scored == lost)
    losses = sum(1 for scored, lost in games if scored < lost)
    points = wins * 3 + draws
    goal_diff = sum(scored - lost for scored, lost in games)
    
    return {
        'wins': wins,
        'draws': draws,
        'losses': losses,
        'points': points,
        'goal_difference': goal_diff
    }

# Example usage
games = [(2, 1), (1, 1), (3, 0), (0, 2), (2, 1)]
stats = calculate_team_stats(games)

print("Team Statistics:")
for key, value in stats.items():
    print(f"  {key}: {value}")
```
