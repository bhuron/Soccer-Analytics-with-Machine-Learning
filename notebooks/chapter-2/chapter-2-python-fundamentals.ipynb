### Sets: Working with Unique Elements

````python
players_scored_last_game = {"Lionel Messi", "Cristiano Ronaldo"}

players_received_red_card = {"Cristiano Ronaldo", "Luis Suárez"}
````

#### Set Operations

````python
all_unique_players = sorted(
    players_scored_last_game | players_received_red_card
)

players_scored_not_red_card = sorted(
    players_scored_last_game - players_received_red_card
)

players_scored_and_red_card = sorted(
    players_scored_last_game & players_received_red_card
)

print(f"All unique players: {all_unique_players}")
print(
    f"Players who scored but did not receive a red card: "
    f"{players_scored_not_red_card}"
)
print(
    f"Players who scored and received a red card: "
    f"{players_scored_and_red_card}"
)
````

### Choosing the Right Data Structure

| Structure | Ordered | Mutable | Duplicates | Use Case |
|-----------|---------|---------|------------|----------|
| List | Yes | Yes | Yes | Sequences, time series |
| Tuple | Yes | No | Yes | Fixed records, coordinates |
| Dictionary | Yes* | Yes | No (keys) | Labeled data, statistics |
| Set | No | Yes** | No | Unique items, comparisons |

*As of Python 3.7+
**Items themselves are immutable

## Conditional Statements: Writing Decision Rules

### Example: Calculating Match Points

````python
team_goals = 2
opponent_goals = 1

if team_goals > opponent_goals:
    points = 3
    result = "Win"
elif team_goals == opponent_goals:
    points = 1
    result = "Draw"
else:
    points = 0
    result = "Loss"

print(f"Result: {result}, Points: {points}")
````

## Loops: Repeating Computation Efficiently

### Using zip() to Process Multiple Lists

````python
team_goals = [2, 1, 3]
opponent_goals = [1, 2, 1]
total_points = 0

for goals_scored, goals_conceded in zip(team_goals, opponent_goals):
    if goals_scored > goals_conceded:
        total_points += 3
    elif goals_scored == goals_conceded:
        total_points += 1

print(f"Total standing points: {total_points}")
````

## Functions: Reusable Logic for Analysis

### Example: Calculate Points Function

````python
def calculate_points(games: list) -> int:
    """Return total points from (goals_scored, goals_conceded) pairs."""
    total_points = 0

    for goals_scored, goals_conceded in games:
        if goals_scored > goals_conceded:
            total_points += 3
        elif goals_scored == goals_conceded:
            total_points += 1

    return total_points

games = [(2, 1), (1, 1), (3, 0), (0, 2)]
total_points = calculate_points(games)
print(f"Total points accumulated: {total_points}")
````

### Example: Goal Difference Function

````python
def calculate_goal_difference(games: list) -> int:
    """Return total goal difference from match results."""
    return sum(
        goals_scored - goals_conceded
        for goals_scored, goals_conceded in games
    )

games = [(2, 1), (1, 1), (3, 0), (0, 2), (2, 1)]
goal_diff = calculate_goal_difference(games)
print(f"Goal difference: {goal_diff:+d}")
````

### Example: Team Statistics Function

````python
def calculate_team_stats(games: list) -> dict:
    """Return wins, draws, losses, points, and goal difference."""
    wins = sum(1 for scored, conceded in games if scored > conceded)
    draws = sum(1 for scored, conceded in games if scored == conceded)
    losses = sum(1 for scored, conceded in games if scored < conceded)
    points = wins * 3 + draws
    goal_difference = sum(
        scored - conceded
        for scored, conceded in games
    )
    
    return {
        'wins': wins,
        'draws': draws,
        'losses': losses,
        'points': points,
        'goal_difference': goal_difference
    }

games = [(2, 1), (1, 1), (3, 0), (0, 2), (2, 1)]
stats = calculate_team_stats(games)

print("Team Statistics:")
for key, value in stats.items():
    print(f"  {key}: {value}")
````
