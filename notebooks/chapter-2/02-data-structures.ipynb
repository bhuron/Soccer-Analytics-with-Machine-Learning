## 1. Lists: Ordered and Changeable Collections

Lists are perfect for storing sequences like goals per game, match results, or player statistics.

```python
# Create a list of goals scored per game
goals_per_game = [0, 1, 1, 0, 2]

print(f"Goals per game: {goals_per_game}")
print(f"Number of games: {len(goals_per_game)}")
print(f"Maximum goals in a game: {max(goals_per_game)}")
print(f"Minimum goals in a game: {min(goals_per_game)}")
```

### List Indexing and Slicing

Remember: Python uses 0-based indexing (first element is at index 0).

```python
# Accessing elements
print(f"First game goals: {goals_per_game[0]}")
print(f"Last game goals: {goals_per_game[-1]}")

# Slicing: [start:end] (end is exclusive)
first_three_games = goals_per_game[0:3]
print(f"First three games: {first_three_games}")
```

### Modifying Lists

```python
# Lists are mutable - we can change them
goals_per_game.append(3)  # Add a new game
print(f"After adding a game: {goals_per_game}")

goals_per_game[0] = 1  # Correct the first game's score
print(f"After correction: {goals_per_game}")
```

## 2. Tuples: Immutable Records

Tuples are great for data that shouldn't change, like match results or historical records.

```python
# Match result: (home_goals, away_goals)
match_result = (2, 1)
print(f"Match result: {match_result}")

# Points record: (wins, losses, draws)
points_record = (3, 0, 1)
print(f"Points record: {points_record}")
```

```python
# Tuples are immutable - this will cause an error
try:
    match_result[0] = 4
except TypeError as e:
    print(f"Error: {e}")
    print("Tuples cannot be modified after creation!")
```

## 3. Dictionaries: Key-Value Pairs

Dictionaries are perfect for storing structured data with named fields.

```python
# Team statistics dictionary
teamAStats = {
    'average_goals_made': 2.5,
    'average_goals_lost': 1.2,
    'average_winning_rate': 0.75,
    'current_standing': 1,
    'points': 45
}

print(f"Team A's average goals made: {teamAStats['average_goals_made']}")
print(f"Team A's current standing: {teamAStats['current_standing']}")
```

### Nested Dictionaries: League Statistics

```python
# Dictionary of dictionaries for multiple teams
league_stats = {
    'Team A': {
        'average_goals_made': 2.5,
        'average_goals_lost': 1.2,
        'points': 45
    },
    'Team B': {
        'average_goals_made': 1.8,
        'average_goals_lost': 1.5,
        'points': 38
    },
    'Team C': {
        'average_goals_made': 1.6,
        'average_goals_lost': 1.7,
        'points': 33
    }
}

# Iterate through teams
for team, stats in league_stats.items():
    gpm = stats['average_goals_made']
    print(f"{team}: {gpm:.2f} average goals made per match")
```

## 4. Sets: Unique Collections

Sets are useful for tracking unique items and performing set operations.

```python
# Players who scored in the last game
players_scored_last_game = {"Lionel Messi", "Cristiano Ronaldo"}

# Players who received a red card
players_received_red_card = {"Cristiano Ronaldo", "Luis Suárez"}

print(f"Scored: {players_scored_last_game}")
print(f"Red cards: {players_received_red_card}")
```

### Set Operations

```python
# Union: All unique players (scored OR red card)
all_unique_players = players_scored_last_game | players_received_red_card
print(f"All unique players: {all_unique_players}")

# Difference: Scored but no red card
players_scored_not_red_card = players_scored_last_game - players_received_red_card
print(f"Scored but no red card: {players_scored_not_red_card}")

# Intersection: Scored AND red card
players_scored_and_red_card = players_scored_last_game & players_received_red_card
print(f"Scored and red card: {players_scored_and_red_card}")
```

## 5. Choosing the Right Data Structure

| Structure | Ordered | Mutable | Duplicates | Use Case |
|-----------|---------|---------|------------|----------|
| List | Yes | Yes | Yes | Sequences, time series |
| Tuple | Yes | No | Yes | Fixed records, coordinates |
| Dictionary | Yes* | Yes | No (keys) | Labeled data, statistics |
| Set | No | Yes** | No | Unique items, comparisons |

*As of Python 3.7+
**Items themselves are immutable
