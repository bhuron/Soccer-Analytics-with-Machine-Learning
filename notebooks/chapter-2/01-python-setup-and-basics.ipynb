## 2. Primitive Data Types

Python has four fundamental primitive data types that form the building blocks of soccer analytics:
- **Integers**: Whole numbers (e.g., goals scored)
- **Floats**: Decimal numbers (e.g., expected goals)
- **Strings**: Text data (e.g., player names)
- **Booleans**: True/False values (e.g., did the team win?)

````python
# Check data types using type()
print(type(True))      # Boolean
print(type(1))         # Integer
print(type(1.0))       # Float
print(type('hello'))   # String
````

### Soccer Analytics Example: Match Data

````python
# Representing match data with primitive types
goals_scored = 3           # Integer
expected_goals = 2.7       # Float
team_name = "Barcelona"    # String
won_match = True           # Boolean

print(f"Team: {team_name}")
print(f"Goals: {goals_scored}")
print(f"xG: {expected_goals}")
print(f"Won: {won_match}")
````

## 3. Basic Operations

### Boolean Operations

````python
# Boolean operations (note: 1 is equivalent to True, 0 to False)
print(1 and True)   # True
print(0 or 1)       # 1
````

### Comparisons

````python
# Comparison operations
print(5 > 1)        # True
print(1 > True)     # False (1 equals True)
print(3 == 3)       # True
print(2 != 3)       # True
````

### Mathematical Operations

````python
# Math operations
print(1 + 5)        # Addition: 6
print(True * 5)     # Multiplication: 5 (True = 1)
print(5 / 2)        # Division: 2.5
print(7 ** 2)       # Exponentiation: 49
````

### String Operations

````python
# String concatenation
result = "Team A" + ":" + "Win"
print(result)  # 'Team A:Win'
````

### Soccer Analytics Example: Goal Difference

````python
# Calculate goal difference
goals_for = 3
goals_against = 1
goal_difference = goals_for - goals_against

print(f"Goal Difference: {goal_difference}")
print(f"Positive goal difference: {goal_difference > 0}")
````

## 4. Variable Assignment

Variables allow us to store and reuse values throughout our analysis.

````python
# Variable assignment
homeTeamGoal = 2
awayTeamGoal = 1

print(homeTeamGoal)
print(awayTeamGoal)
````

### Naming Conventions

Two popular conventions:
- **camelCase**: First word lowercase, subsequent words capitalized (e.g., `homeTeamGoal`)
- **snake_case**: All lowercase with underscores (e.g., `home_team_goal`)

Choose one and be consistent!

````python
# snake_case example
home_team_goal = 2
away_team_goal = 1

print(home_team_goal)
print(away_team_goal)
````
