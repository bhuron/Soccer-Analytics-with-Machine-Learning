## What is Classification?

Classification is about **sorting** things into categories. Instead of predicting a continuous number (like regression), we predict a **discrete label** or **class**.

Think about these soccer questions:
- Will this shot be a **goal** or **not**?
- Will this pass be **completed** or **intercepted**?
- Will the match end in a **win**, **draw**, or **loss**?

These are all classification problems!

## Binary vs. Multi-Class Classification

### Binary Classification

Binary classification has **two possible outcomes**. Examples:

1. **Goal or No Goal?** - Basis for Expected Goals (xG) models
2. **Pass Completed or Incomplete?** - Evaluating passing ability  
3. **Successful Tackle or Failed Tackle?** - Assessing defensive skill

### Multi-Class Classification

Multi-class classification has **three or more categories**. Examples:

1. **Match Outcome**: Win, Draw, or Loss
2. **Pass Type**: Short Pass, Long Ball, or Cross
3. **Shot Type**: Open Play, Free Kick, or Penalty

We'll start with binary classification as it's simpler to understand.

## Real-World Example: Shot Outcomes

Let's load some real soccer data and look at shot outcomes.

```python
# Load Champions League Final 2019 data
matches = sb.matches(competition_id=16, season_id=4)
cl_final = matches[matches['match_id'] == 22912]

print("Match Information:")
print(f"{cl_final['home_team'].values[0]} vs {cl_final['away_team'].values[0]}")
print(f"Score: {cl_final['home_score'].values[0]}-{cl_final['away_score'].values[0]}")
```

```python
# Get event data
events = sb.events(match_id=22912)

# Filter for shots
shots = events[events['type'] == 'Shot'].copy()

print(f"\nTotal shots in the match: {len(shots)}")
print(f"\nShot outcomes:")
print(shots['shot_outcome_name'].value_counts())
```

**What we see:**
- Most shots are **not goals** (Saved, Blocked, Off Target)
- Only a few shots result in **Goal**
- This is a **binary classification** problem: Goal vs No Goal
- The classes are **imbalanced** (more non-goals than goals)

## Visualizing the Classification Problem

```python
# Create binary target variable
shots['is_goal'] = (shots['shot_outcome_name'] == 'Goal').astype(int)

# Visualize class distribution
plt.figure(figsize=(8, 5))
shots['is_goal'].value_counts().plot(kind='bar')
plt.xticks([0, 1], ['No Goal', 'Goal'], rotation=0)
plt.ylabel('Count')
plt.title('Shot Outcomes: Binary Classification')
plt.tight_layout()
plt.show()

print(f"\nClass distribution:")
print(f"No Goal: {(shots['is_goal'] == 0).sum()} ({(shots['is_goal'] == 0).mean()*100:.1f}%)")
print(f"Goal: {(shots['is_goal'] == 1).sum()} ({(shots['is_goal'] == 1).mean()*100:.1f}%)")
```

## The Classification Workflow

Every classification project follows these steps:

1. **Define the Problem** - What are we trying to predict?
2. **Gather Data** - Collect relevant features and outcomes
3. **Prepare Data** - Clean, transform, create features
4. **Choose Algorithm** - Select appropriate classifier
5. **Train Model** - Learn patterns from training data
6. **Evaluate Performance** - Test on unseen data
7. **Interpret Results** - Understand what the model learned

We'll follow this workflow throughout Chapter 4.
