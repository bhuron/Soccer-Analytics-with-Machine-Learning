# 1. Types of Soccer Bets

Soccer betting offers various options, each with different risk-reward profiles.

## 1.1 Match Result (1X2)

The **1X2 bet** is the simplest and most popular form. You predict one of three outcomes:

- **1**: Home team wins
- **X**: Draw
- **2**: Away team wins

### Example

**Manchester City vs. Liverpool**
- 1 (Man City wins): 2.00 odds
- X (Draw): 3.50 odds
- 2 (Liverpool wins): 4.00 odds

### Why 1X2?

- Most **liquid market** (highest betting volume)
- Odds are typically closest to "true" probabilities
- Foundation for many betting strategies

## 1.2 Over/Under (Totals)

Predict whether total goals will be **above or below** a threshold.

### Example

**Over/Under 2.5 goals**
- Over 2.5: Win if match has 3+ goals
- Under 2.5: Win if match has 0-2 goals

### When to Use

- You have insights about **team tactics** (defensive vs offensive)
- **Weather conditions** might affect scoring
- Teams are known for high/low-scoring matches

## 1.3 Both Teams to Score (BTTS)

Predict whether **both teams will score at least one goal**.

### Example

- BTTS Yes: Both teams score
- BTTS No: At least one team doesn't score

### Appeal

- Outcome is **independent of final result**
- Keeps match interesting regardless of winner
- Popular among casual bettors

## 1.4 Correct Score

Predict the **exact final score** of the match.

### Example

- 2-1 Home Win
- 0-0 Draw
- 1-3 Away Win

### Characteristics

- **Highest payouts** (long odds)
- **Notoriously difficult** to predict accurately
- Requires both score and outcome to be correct

# 2. Understanding Odds

Odds are the **backbone of sports betting**. They:
- Determine potential payout
- Reflect bookmaker's assessment of probability
- Come in three main formats

## 2.1 Decimal Odds (European)

**Most straightforward format**, widely used in Europe and Australia.

Represents **total return** on a unit stake (including original wager).

### Formula

```
Total Return = Stake × Decimal Odds
Profit = (Stake × Decimal Odds) - Stake
```

### Example

- Odds: 2.50
- $10 bet returns: $10 × 2.50 = **$25**
- Profit: $25 - $10 = **$15**

````python
def decimal_payout(stake, odds):
    """
    Calculate total return and profit for decimal odds.
    """
    total_return = stake * odds
    profit = total_return - stake
    return total_return, profit

# Example
stake = 10
odds = 2.50

total, profit = decimal_payout(stake, odds)
print(f"Decimal Odds: {odds}")
print(f"Stake: ${stake}")
print(f"Total Return: ${total:.2f}")
print(f"Profit: ${profit:.2f}")
````

## 2.2 Fractional Odds (UK)

Commonly used in the UK, shows **potential profit relative to stake**.

### Formula

```
Profit = Stake × (Numerator / Denominator)
Total Return = Stake + Profit
```

### Example

- Odds: 3/1 ("three to one")
- $10 bet yields: $10 × (3/1) = **$30 profit**
- Total return: $10 + $30 = **$40**

````python
def fractional_payout(stake, numerator, denominator):
    """
    Calculate total return and profit for fractional odds.
    """
    profit = stake * (numerator / denominator)
    total_return = stake + profit
    return total_return, profit

# Example
stake = 10
numerator, denominator = 3, 1  # 3/1 odds

total, profit = fractional_payout(stake, numerator, denominator)
print(f"Fractional Odds: {numerator}/{denominator}")
print(f"Stake: ${stake}")
print(f"Total Return: ${total:.2f}")
print(f"Profit: ${profit:.2f}")
````

## 2.3 Moneyline Odds (American)

Popular in the US, uses **positive or negative numbers**.

### Positive Moneyline (+)

Shows profit on a **$100 stake**.

- +150: $100 stake yields **$150 profit**
- Total return: $250

### Negative Moneyline (-)

Shows stake needed to profit **$100**.

- -150: Must stake **$150** to profit $100
- Total return: $250

````python
def moneyline_payout(stake, moneyline):
    """
    Calculate total return and profit for moneyline odds.
    """
    if moneyline > 0:
        # Positive moneyline
        profit = stake * (moneyline / 100)
    else:
        # Negative moneyline
        profit = stake * (100 / abs(moneyline))
    
    total_return = stake + profit
    return total_return, profit

# Example: Positive moneyline
stake = 100
moneyline = +150

total, profit = moneyline_payout(stake, moneyline)
print(f"Moneyline Odds: {moneyline:+d}")
print(f"Stake: ${stake}")
print(f"Total Return: ${total:.2f}")
print(f"Profit: ${profit:.2f}")

print("\n" + "="*40 + "\n")

# Example: Negative moneyline
stake = 150
moneyline = -150

total, profit = moneyline_payout(stake, moneyline)
print(f"Moneyline Odds: {moneyline:+d}")
print(f"Stake: ${stake}")
print(f"Total Return: ${total:.2f}")
print(f"Profit: ${profit:.2f}")
````

## 2.4 Converting Between Odds Formats

Let's create functions to convert between all three formats.

````python
def decimal_to_fractional(decimal_odds):
    """
    Convert decimal odds to fractional odds.
    """
    profit_ratio = decimal_odds - 1
    # Simplify to nearest common fraction
    from fractions import Fraction
    frac = Fraction(profit_ratio).limit_denominator(100)
    return frac.numerator, frac.denominator

def decimal_to_moneyline(decimal_odds):
    """
    Convert decimal odds to moneyline odds.
    """
    if decimal_odds >= 2.0:
        return int((decimal_odds - 1) * 100)
    else:
        return int(-100 / (decimal_odds - 1))

def fractional_to_decimal(numerator, denominator):
    """
    Convert fractional odds to decimal odds.
    """
    return 1 + (numerator / denominator)

def moneyline_to_decimal(moneyline):
    """
    Convert moneyline odds to decimal odds.
    """
    if moneyline > 0:
        return 1 + (moneyline / 100)
    else:
        return 1 + (100 / abs(moneyline))

# Example conversions
print("Odds Conversion Examples:\n")

decimal = 2.50
frac_num, frac_den = decimal_to_fractional(decimal)
moneyline = decimal_to_moneyline(decimal)

print(f"Decimal: {decimal}")
print(f"  → Fractional: {frac_num}/{frac_den}")
print(f"  → Moneyline: {moneyline:+d}")

print("\n" + "="*40 + "\n")

frac_num, frac_den = 3, 1
decimal = fractional_to_decimal(frac_num, frac_den)
moneyline = decimal_to_moneyline(decimal)

print(f"Fractional: {frac_num}/{frac_den}")
print(f"  → Decimal: {decimal:.2f}")
print(f"  → Moneyline: {moneyline:+d}")
````

## 2.5 Visualizing Odds Relationships

````python
# Create range of decimal odds
odds_range = np.linspace(1.1, 10, 100)

# Calculate payouts for $10 stake
payouts = [(stake * odds, (stake * odds) - stake) for odds in odds_range]
total_returns = [p[0] for p in payouts]
profits = [p[1] for p in payouts]

stake = 10

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left panel: Odds vs Payout
ax1.plot(odds_range, total_returns, linewidth=2, color='steelblue', label='Total Return')
ax1.plot(odds_range, profits, linewidth=2, color='coral', label='Profit')
ax1.axhline(stake, color='black', linestyle='--', alpha=0.5, label=f'Stake (${stake})')
ax1.set_xlabel('Decimal Odds', fontsize=12)
ax1.set_ylabel('Amount ($)', fontsize=12)
ax1.set_title(f'Payout vs Odds (${stake} stake)', fontsize=14, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Right panel: Odds comparison for different stakes
stakes = [10, 20, 50, 100]
for stake_val in stakes:
    profits = [(stake_val * odds) - stake_val for odds in odds_range]
    ax2.plot(odds_range, profits, linewidth=2, label=f'${stake_val} stake')

ax2.set_xlabel('Decimal Odds', fontsize=12)
ax2.set_ylabel('Profit ($)', fontsize=12)
ax2.set_title('Profit vs Odds for Different Stakes', fontsize=14, fontweight='bold')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Notice how profit scales linearly with both odds and stake!")
````

# 3. Calculating Implied Probabilities

**Implied probability** is a key concept representing the likelihood of an outcome as indicated by the odds.

It's essential for:
- Understanding the bookmaker's perspective
- Identifying **value bets**
- Comparing your model's predictions to the market

## 3.1 Formula for Decimal Odds

```
Implied Probability = 1 / Decimal Odds
```

### Example

**Premier League Match**

| Outcome | Odds | Implied Probability |
|---------|------|--------------------|
| Home Win | 2.00 | 50.0% |
| Draw | 3.50 | 28.6% |
| Away Win | 4.00 | 25.0% |

````python
def odds_to_probability(decimal_odds):
    """
    Convert decimal odds to implied probability.
    """
    return 1 / decimal_odds

# Example match
match_odds = {
    'Home Win': 2.00,
    'Draw': 3.50,
    'Away Win': 4.00
}

print("Implied Probabilities:\n")
for outcome, odds in match_odds.items():
    prob = odds_to_probability(odds)
    print(f"{outcome:12s}: {odds:.2f} odds → {prob:.1%} probability")

# Calculate total
total_prob = sum(odds_to_probability(odds) for odds in match_odds.values())
print(f"\nTotal: {total_prob:.1%}")
print(f"\n⚠️  Notice: Total exceeds 100%! This is the 'overround' (bookmaker's margin)")
````

## 3.2 The Overround (Bookmaker's Margin)

The **overround** is the bookmaker's built-in profit margin.

### Why does the total exceed 100%?

Bookmakers adjust odds to ensure they make a profit regardless of the outcome.

### Formula

```
Overround = (Sum of Implied Probabilities) - 1
Margin % = Overround × 100
```

````python
def calculate_overround(odds_list):
    """
    Calculate overround (bookmaker's margin) from a list of odds.
    """
    total_prob = sum(1/odds for odds in odds_list)
    overround = total_prob - 1
    return overround, overround * 100

# Calculate for our example
odds_list = list(match_odds.values())
overround, margin_pct = calculate_overround(odds_list)

print(f"Bookmaker's Overround: {overround:.4f}")
print(f"Bookmaker's Margin: {margin_pct:.2f}%")
print(f"\nInterpretation: The bookmaker expects to make {margin_pct:.2f}% profit on this market.")
````

## 3.3 Fair Odds (Removing the Overround)

To get **fair odds** (true probabilities), we need to remove the bookmaker's margin.

````python
def calculate_fair_odds(odds_dict):
    """
    Calculate fair odds by removing the overround.
    """
    # Calculate total implied probability
    total_prob = sum(1/odds for odds in odds_dict.values())
    
    # Calculate fair probabilities
    fair_probs = {outcome: (1/odds) / total_prob 
                  for outcome, odds in odds_dict.items()}
    
    # Convert back to odds
    fair_odds = {outcome: 1/prob for outcome, prob in fair_probs.items()}
    
    return fair_probs, fair_odds

# Calculate fair odds
fair_probs, fair_odds = calculate_fair_odds(match_odds)

print("Comparison: Bookmaker Odds vs Fair Odds\n")
print(f"{'Outcome':<12} {'Book Odds':<12} {'Fair Odds':<12} {'Book Prob':<12} {'Fair Prob'}")
print("="*65)

for outcome in match_odds:
    book_odds = match_odds[outcome]
    fair_odd = fair_odds[outcome]
    book_prob = 1/book_odds
    fair_prob = fair_probs[outcome]
    
    print(f"{outcome:<12} {book_odds:<12.2f} {fair_odd:<12.2f} {book_prob:<12.1%} {fair_prob:.1%}")

print("\n✅ Fair probabilities now sum to exactly 100%!")
print(f"   Total: {sum(fair_probs.values()):.1%}")
````

## 3.4 Visualizing Odds to Probability Conversion

````python
# Create range of odds
odds_range = np.linspace(1.01, 20, 200)
probabilities = [1/odds for odds in odds_range]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left panel: Odds to Probability
ax1.plot(odds_range, probabilities, linewidth=2, color='darkblue')
ax1.set_xlabel('Decimal Odds', fontsize=12)
ax1.set_ylabel('Implied Probability', fontsize=12)
ax1.set_title('Odds to Probability Conversion', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axhline(0.5, color='red', linestyle='--', alpha=0.5, label='50% (Even odds)')
ax1.axvline(2.0, color='red', linestyle='--', alpha=0.5)
ax1.legend(fontsize=11)

# Add annotations
ax1.annotate('Odds = 2.0\nProb = 50%', xy=(2.0, 0.5), xytext=(5, 0.6),
            arrowprops=dict(arrowstyle='->', color='red', lw=1.5),
            fontsize=10, color='red')

# Right panel: Payout comparison
stake = 10
odds_examples = [1.5, 2.0, 3.0, 5.0, 10.0]
profits = [(odds - 1) * stake for odds in odds_examples]
probs = [1/odds for odds in odds_examples]

x = np.arange(len(odds_examples))
width = 0.35

ax2_twin = ax2.twinx()

bars1 = ax2.bar(x - width/2, profits, width, label='Profit ($)', color='green', alpha=0.7)
bars2 = ax2_twin.bar(x + width/2, [p*100 for p in probs], width, label='Probability (%)', color='orange', alpha=0.7)

ax2.set_xlabel('Decimal Odds', fontsize=12)
ax2.set_ylabel('Profit on $10 Bet ($)', fontsize=12, color='green')
ax2_twin.set_ylabel('Implied Probability (%)', fontsize=12, color='orange')
ax2.set_title('Payout vs Probability Trade-off', fontsize=14, fontweight='bold')
ax2.set_xticks(x)
ax2.set_xticklabels([f'{o:.1f}' for o in odds_examples])
ax2.tick_params(axis='y', labelcolor='green')
ax2_twin.tick_params(axis='y', labelcolor='orange')
ax2.grid(True, alpha=0.3, axis='y')

# Add legends
ax2.legend(loc='upper left', fontsize=10)
ax2_twin.legend(loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()

print("Key Insight: Higher odds = Higher profit BUT lower probability!")
````

# 4. Value Bets: The Foundation of Profitable Betting

A **value bet** occurs when you believe the bookmaker has underestimated the likelihood of an event.

### Formula

```
Value Bet: Your Probability > Implied Probability from Odds
```

### Example

- Bookmaker odds: 3.00 (implies 33.3% probability)
- Your model predicts: 40% probability
- **This is a value bet!** (40% > 33.3%)

````python
def identify_value_bet(your_probability, bookmaker_odds):
    """
    Identify if a bet has value.
    
    Returns:
        is_value: Boolean indicating if bet has value
        edge: Your edge over the bookmaker (percentage points)
        expected_value: Expected value of a $1 bet
    """
    implied_prob = 1 / bookmaker_odds
    edge = your_probability - implied_prob
    
    # Expected value of a $1 bet
    win_amount = bookmaker_odds - 1  # Profit if win
    expected_value = (your_probability * win_amount) - ((1 - your_probability) * 1)
    
    is_value = edge > 0
    
    return is_value, edge, expected_value

# Example
your_prob = 0.40
bookmaker_odds = 3.00

is_value, edge, ev = identify_value_bet(your_prob, bookmaker_odds)

print(f"Your Probability: {your_prob:.1%}")
print(f"Bookmaker Odds: {bookmaker_odds:.2f}")
print(f"Implied Probability: {1/bookmaker_odds:.1%}")
print(f"\nYour Edge: {edge:+.1%}")
print(f"Expected Value (per $1 bet): ${ev:+.3f}")
print(f"\n{'✅ VALUE BET!' if is_value else '❌ No value'}")

if is_value:
    print(f"\nInterpretation: Over many bets, you expect to make ${ev:.3f} per $1 wagered.")
````
